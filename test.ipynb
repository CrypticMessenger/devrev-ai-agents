```python
from turtle import color
import matplotlib.pyplot as plt

# Data
labels = ['LangChain + React', 'Palm + Segmentation \n+ vector DB', 'Palm + Segmentation']
values = [react_latency, palm_latency_w_embeddings, palm_latency_wo_embeddings,]

# Plotting
fig, ax = plt.subplots()
bars = ax.bar(labels, values, width=0.4, color=['#1f77b4', '#ff7f0e', '#af55a0'], align='center')

# Adding labels and title
ax.set_ylabel('Average Latency (s)')
ax.set_title('Average Latency Comparison')
# ax.grid(axis='y')

# Display the values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# Show the plot
plt.show()
```
